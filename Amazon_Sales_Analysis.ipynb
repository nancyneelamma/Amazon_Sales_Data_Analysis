```python
# Copying the dataset into another variable, to use and make changes

salesdata = data.copy()
print(salesdata)
```

```python
# Dropping irrelavent columns

cols = ["Order Priority","Order ID","Units Sold","Unit Price","Unit Cost"]
dropped_cols = salesdata.drop(cols,axis=1,inplace=True)
print(dropped_cols)
```

```python
# Convert order date and ship date todatetime format and extract month,year,day from it

salesdata['Order Date'] = pd.to_datetime(salesdata['Order Date'])

salesdata['OrderMonth'] = salesdata['Order Date'].dt.month
salesdata['OrderYear'] = salesdata['Order Date'].dt.year
salesdata['OrderDate'] = salesdata['Order Date'].dt.day

salesdata['Ship Date'] = pd.to_datetime(salesdata['Ship Date'])

salesdata['ShipMonth'] = salesdata['Ship Date'].dt.month
salesdata['ShipYear'] = salesdata['Ship Date'].dt.year
salesdata['ShipDate'] = salesdata['Ship Date'].dt.day
```

**Sales trends**

```python
# Month-wise sales trend

sales_monthly = salesdata.groupby('OrderMonth')['Total Revenue'].sum().reset_index()

# Plotting Graph
plt.figure(figsize=(12,6))
sns.barplot(x='OrderMonth',y='Total Revenue',data = sales_monthly, color='Green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.grid(axis='y')
plt.show()
```

```python
# Year-wise sales trend

sales_yearly = salesdata.groupby('OrderYear')['Total Revenue'].sum().reset_index()

# Plotting Graph
plt.figure(figsize=(12,6))
sns.barplot(x='OrderYear',y='Total Revenue',data = sales_yearly, color='purple')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.grid(axis='y')
plt.show()
```

```python
# Yearly_month-wise sales trend

sales_yearly_month = salesdata.groupby(['OrderYear','OrderMonth'])['Total Revenue'].sum().reset_index()
years = sales_yearly_month['OrderYear'].unique()
for year in years:
  yearly_sales = sales_yearly_month[sales_yearly_month['OrderYear']==year]
  # Plotting Graph
  plt.figure(figsize=(8,4))
  sns.lineplot(x='OrderMonth',y='Total Revenue',data = yearly_sales, color='orange', marker='o')
  plt.title(f' Year {year} Monthly Sales Trend')
  plt.xlabel('Months')
  plt.ylabel('Total Sales')
  plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
  plt.xticks(np.arange(1,13))
  plt.grid(True)
  plt.show()
```

**Key metrics and their relationships with attribues**

```python
# Total Sales and Region
regional_sales = salesdata.groupby('Region')['Total Revenue'].sum().reset_index()
regional_sales_sort = regional_sales.sort_values(by='Total Revenue', ascending=False)

# Plotting Graph
plt.figure(figsize=(14,10))
sns.barplot(x='Region',y='Total Revenue',data = regional_sales_sort , color='lightgreen')
plt.title('Regional Sales Performance')
plt.xlabel('Regions')
plt.ylabel('Total Sales')
plt.xticks(fontsize=6.9)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Total Sales by each Region over the years
regional_sales = salesdata.groupby(['Region','OrderYear'])['Total Revenue'].sum().reset_index()
regional_sales_sort = regional_sales.sort_values(by='Total Revenue', ascending=True)

# Creating custom colors for the regions for Visualization
custom_palette = {'Asia': 'Purple', 'North America': 'Green', 'Australia and Oceania': 'Yellow',
                  'Sub-Saharan Africa': 'Gray','Europe': 'Skyblue','Middle East and North Africa': 'Brown','Central America and the Caribbean': 'Red'}
# Plotting Graph
plt.figure(figsize=(10,6))
sns.barplot(x='OrderYear',y='Total Revenue',data = regional_sales_sort , hue='Region',palette=custom_palette)

plt.title('Total Sales by each Region over the years')
plt.xlabel('Years')
plt.ylabel('Total Sales')
plt.legend(title='Region')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.tight_layout()
plt.show()
```

```python
# Total Profit and Country

country = data.groupby('Country')['Total Profit'].sum().reset_index()
country_sort = country.sort_values(by='Total Profit',ascending = False)

# Plotting Graph
plt.figure(figsize=(14,8))
sns.barplot(x='Total Profit',y='Country',data=country_sort, color='Brown')
plt.title('Profits by each Country')
plt.xlabel('Total Profit')
plt.ylabel('Countries')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.yticks(fontsize=5.4)
plt.show()
```

```python
# Comparing Top 10 countries having maximum sales and profits

countries = salesdata.groupby('Country').agg(Total_Sales=('Total Revenue','sum'),Total_Profit=('Total Profit','sum')).reset_index()

# Top 10 countries by sales and profit
top_10_countries_sales = countries.nlargest(10, 'Total_Sales')
top_10_countries_profit = countries.nlargest(10, 'Total_Profit')

# Plotting Graph
plt.figure(figsize=(10,6))
plt.bar(top_10_countries_sales['Country'],top_10_countries_sales['Total_Sales'],color='skyblue',label='Total Sales')
plt.bar(top_10_countries_profit['Country'],top_10_countries_profit['Total_Profit'],color='lightgreen',label='Total Profit')
plt.title('Top 10 Countries having higher Sales and Profit')
plt.xlabel('Countries')
plt.ylabel('Sales & Profit')
plt.xticks(fontsize=9,rotation=45)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Comparing 10 countries having lowest sales and profits

countries = salesdata.groupby('Country').agg(Total_Sales=('Total Revenue','sum'),Total_Profit=('Total Profit','sum')).reset_index()

# Top 10 countries by sales and profit
countries10_sales = countries.nsmallest(10, 'Total_Sales')
countries10_profit = countries.nsmallest(10, 'Total_Profit')

# Plotting Graph
plt.figure(figsize=(10,6))
plt.bar(countries10_sales['Country'],countries10_sales['Total_Sales'],color='skyblue',label='Total Sales')
plt.bar(countries10_profit['Country'],countries10_profit['Total_Profit'],color='lightgreen',label='Total Profit')
plt.title('Top 10 Countries having lowest Sales and Profit')
plt.xlabel('Countries')
plt.ylabel('Sales & Profit')
plt.xticks(fontsize=9,rotation=45)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Total Sales and Item Type

item_types = salesdata.groupby('Item Type')['Total Revenue'].sum().reset_index()
item_types_sort = item_types.sort_values(by='Total Revenue', ascending=False)

# Plotting Graph
plt.figure(figsize=(10,6))
sns.barplot(x='Item Type',y='Total Revenue',data = item_types_sort, color='skyblue')
plt.title('Items')
plt.xlabel('Item Type ')
plt.ylabel('Total Sales')
plt.xticks(fontsize=6.9)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.grid(axis='y')
plt.show()
```

```python
# Items in the regions with profits

# Calculate total profit for each item in each region

profits_regions_items = salesdata.groupby(['Region','Item Type'])['Total Profit'].sum().reset_index()

# Regions with profit

profits_regions = profits_regions_items.groupby('Region').filter(lambda x:x['Total Profit'].sum() > 0)

# Plot the items for each region with profits

for region,items in profits_regions_items.groupby('Region'):
  plt.figure(figsize=(10,6))
  sns.barplot(x='Item Type',y='Total Profit',data=items, color='gray')
  plt.title(f'Items in {region} by Total Sales')
  plt.xlabel('Items')
  plt.ylabel('Total Profit')
  plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
  plt.xticks(rotation=45)
  plt.grid(axis='y')
  plt.show()
```

```python
# Calculating Total Sales for Sales Channel

sales_channel = salesdata.groupby('Sales Channel')['Total Revenue'].sum().reset_index()
sales_channel_sort = sales_channel.sort_values(by='Total Revenue',ascending = False)

# Plotting Graph
plt.figure(figsize=(8,6))
sns.barplot(x='Sales Channel',y='Total Revenue',data=sales_channel_sort,color='lightcoral')
plt.title('Sales Channel Effectiveness')
plt.xlabel('Sales Channel')
plt.ylabel('Total Sales')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,_: '{:.0f}'.format(x)))
plt.xticks(fontsize=9)
plt.show()
```

```python
# Calculating Total Sales,Total Profits and Average Order Processing time for each items and sales channel

salesdata['Order Processing Time'] = salesdata['ShipDate'] - salesdata['OrderDate']
ordertime_itemtype_saleschannel = salesdata.groupby(['Item Type','Sales Channel']).agg(avg_order_process_time=('Order Processing Time','mean'),
                                  Total_Sales=('Total Revenue','sum'),Total_Profit=('Total Profit','sum')).reset_index()

# Pivotting the data
pivot_data = ordertime_itemtype_saleschannel.pivot(index='Item Type',columns='Sales Channel')

#Plotting heatmap
plt.figure(figsize=(20,8))
sns.heatmap(pivot_data, annot=True, cmap='coolwarm',fmt='.2f',linewidth=1,linecolor='black')
plt.title('Average Processing Time and Total Sales by Item and Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Items')
plt.xticks(fontsize=9,rotation=360)
plt.show()
```
